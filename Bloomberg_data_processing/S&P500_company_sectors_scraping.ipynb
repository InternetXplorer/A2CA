```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
wikiurl="https://en.wikipedia.org/wiki/List_of_S%26P_500_companies?fbclid=IwAR20RhvspvTHghzmyOOnfQYTJjhfWyt6nbdM9iJu31fYejkvWXhoGz6F97E"
table_class="wikitable sortable jquery-tablesorter"
response=requests.get(wikiurl)
print(response.status_code)
```

```python
# parse data from the html into a beautifulsoup object
soup = BeautifulSoup(response.text, 'html.parser')
indiatable=soup.find('table',{'class':"wikitable"})
```

```python
df=pd.read_html(str(indiatable))
# convert list to dataframe
df=pd.DataFrame(df[0])
df
```

```python
df=df.iloc[:,0:6]
df.drop("SEC filings",axis=1,inplace=True)
```

```python
df.rename(columns={"Symbol":"Ticker"},inplace=True)
df.head()
```

```python
df.to_csv('S&P500_company_sectors.csv', sep=",", index=False)
```

```python
sp500=pd.read_csv("preprocessed_sp500_data_01-02-22.csv")
for i in range(len(sp500)):
    sp500.loc[i,"Ticker"] = sp500.loc[i,"Ticker"].split()[0] #extract ticker from long ticker code

sp500=df.merge(sp500,on='Ticker', how='left') #merge both dataframes on ticker column
sp500.head()
sp500.to_csv('preprocessed_sp500_data_01-02-22_with_sectors.csv', sep=",", index=False)
```
