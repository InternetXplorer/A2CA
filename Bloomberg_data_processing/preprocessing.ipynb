### Imports

```python
import pandas as pd
```

### Data loading

```python
path = "original_data/"
filename = "sp500_data_01-02-22.csv"

data = pd.read_csv(path + filename, sep=";", skiprows=1, decimal=",", na_values=['#N/A Field Not Applicable'])
```

```python
data.head(2)
```

```python
data.info()
```

### Dropping useless column & renaming

```python
data = data.rename(columns={data.columns.values[0]:"Ticker"})

date_column = "Datetime of data import"
if(date_column in data.columns.values):
    data = data.drop(columns=["Datetime of data import"])
```

```python
data.head(2)
```

### Show some statistics

```python
print("Number of missing values per column :\n")
print(data.isnull().sum())
```

### Save to CSV file

```python
new_filename = "preprocessed_" + filename
data.to_csv(new_filename, sep=",", index=False)
```
